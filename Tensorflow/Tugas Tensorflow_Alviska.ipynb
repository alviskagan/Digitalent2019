```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

```
# x1 = 0.7
# x2 = 2.1

x1 = tf.constant([0.7])
x2 = tf.constant([2.1])

# w1 = 0.5
# w2 = -0.3
w1 = tf.constant([0.5])
w2 = tf.constant([-0.3])

# b = 1.2
b = tf.constant([1.2])
```

```
# net = b + (w1*x1)+(w2*x2)
w1x1 = tf.multiply(w1,x1)
w2x2 = tf.multiply(w2,x2)

wx = tf.add(w1x1,w2x2)

# net = tf.add(b,wx)
```

```
add_bias_OP = tf.add(wx, b, name="add_bias") 
activation_OP = tf.nn.sigmoid(add_bias_OP, name="activation")
```

```
with tf.Session() as session:
    net = session.run(activation_OP)
    print('nilai f(net) = ',net)
    if net < 0:
      print("nilai Y = ", 0)
    else:
      print("nilai Y = ", 1)
```
